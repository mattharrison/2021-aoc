```python
import numpy as np
import pandas as pd
df = pd.read_csv('day2.txt', sep=' ', names=['direction', 'amount'])
x = df[df.direction == 'forward'].amount.sum()
y = df[df.direction == 'down'].amount.sum() - df[df.direction == 'up'].amount.sum()
x, y, x*y
```

```python
(df
 .assign(aim_delta=pd.Series(np.select([df.direction == 'down', df.direction == 'up'], [df.amount, -df.amount])),
         aim=lambda df_: df_.aim_delta.cumsum(),
         x=lambda df_: df_.amount.where(df_.direction=='forward').cumsum().ffill().astype(int),
         y=lambda df_: (df_.amount.where(df_.direction=='forward')*df_.aim).cumsum().ffill().astype(int),
         x_y = lambda df_:df_.x*df_.y
        )
)
#df[df.direction == 'down'].amount - df[df.direction == 'up'].amount
```
